## Matplotlib를 이용한 시각화 기초

#### matplotlib란?

- 플롯(그래프)을 그릴 때 주로 쓰이는 2D, 3D 플롯팅 패키지

### 선(line)으로 시각화하기

#### Python package 가져오기 및 matplotlib 출력 옵션 설정

- matplotlib.pyplot : matplotlib의 서브패키지로 플롯을 그려주는 패키지.  
- numpy : 행렬, 벡터 등의 수학 계산을 위한 자료구조와 계산 함수를 제공하는 패키지.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

#### 임의의(랜덤) 값 40개로 이루어진 배열 변수 생성

```python
value = np.random.standard_normal(40)
value
```

#### 선 plot 그리기 및 x/y축 범위 설정

```python
index = range(len(value))

# x, y 축 범위 설정
plt.xlim(0, 20)
plt.ylim(np.min(value) - 1, np.max(value) + 1)
plt.plot(index, value)
```

#### plot 옵션 설정 및 꾸미기

```python
plt.figure(figsize=(7, 4)) # plot의 새로운 figure(모양)을 생성하는 함수
plt.plot(value.cumsum(), 'b', lw=1.5)
plt.plot(value.cumsum(), 'ro')
plt.xlabel('index')
plt.ylabel('value')
plt.title('Line Plot 1')
```

#### 랜덤 값으로 이루어진 2차원 배열 변수 생성하기

```python
value = np.random.standard_normal((30, 2))
value
```

#### 2차원 데이터 생성 및 시각화

- 하나의 플롯 그래프 안에 두 개의 데이터 셋이 독립적으로 그려진다.
- 각각의 데이터 셋에 라벨을 붙여서 구분한다.

```python
plt.figure(figsize=(10, 4))
plt.plot(value[:,0], lw=1.5, label='1st')
plt.plot(value[:,1], lw=1.5, label='2nd')
plt.plot(value, 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
plt.title('Line Plot 2')
```

#### 데이터 별로 각각 그리기

- subplot() 함수를 이용해 두 개의 plot을 배열 형태로 분리한다.

```python
# 1번째
plt.figure(figsize=(10, 5))
plt.subplot(211)
plt.plot(value[:, 0], lw=1.5, label='1st')
plt.plot(value[:, 0], 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('Line Plot 3')

# 2번째
plt.subplot(212)
plt.plot(value[:, 1], lw=1.5, label='2nd')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.ylabel('value')
```

#### 6개의 각각 다른 서브플롯 그려보기

```python
# 1번째
plt.figure(figsize=(13, 5))
plt.subplot(231)
plt.plot(value[:, 0], lw=1.5, label='1st')
plt.plot(value[:, 0], 'co')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('Line Plot 3')

# 2번째
plt.subplot(232)
plt.plot(value[:, 0], 'g-.', lw=1.5, label='1st')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.ylabel('value')

# 3번째
plt.subplot(233)
plt.plot(value[:, 0], 'g', lw=1.5, label='1st')
plt.plot(value[:, 0], 'bD')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.ylabel('value')

# 4번째
plt.subplot(234)
plt.plot(value[:, 1], '*', lw=1.5, label='2nd')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.ylabel('value')

# 5번째
plt.subplot(235)
plt.plot(value[:, 1], 'b', lw=1.5, label='2nd')
plt.plot(value[:, 1], 'ms')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.ylabel('value')

# 6번째
plt.subplot(236)
plt.plot(value[:, 1], 'r--', lw=1.5, label='2nd')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.ylabel('value')
```

#### pyplot.plot() 함수를 이용한 산점도 그리기

```python
value = np.random.standard_normal((500, 2))

plt.plot(value[:,0], value[:,1], 'ro')
plt.grid(False)
plt.xlabel('value 1')
plt.ylabel('value 2')
plt.title('Scatter Plot 1')
```

#### pyploy.scatter() 함수를 이용한 산점도 그리기

```python
plt.figure(figsize=(7, 5))
plt.scatter(value[:,0], value[:,1], marker='o')
plt.grid(True)
plt.xlabel('value 1')
plt.ylabel('value 2')
plt.title('Scatter Plot 2')
```

```python
# 산점도에 색깔 입히기 (c 옵션 사용)
color = np.random.randint(0, 10, len(value))

plt.figure(figsize=(10, 5))
plt.scatter(value[:,0], value[:,1], c=color, marker='o')
plt.colorbar()
plt.xlabel('value 1')
plt.ylabel('value 2')
plt.title('Scatter Plot 3')
```

### 히스토그램(histogram)으로 시각화하기

#### pyplot.hist() 함수를 이용한 히스토그램 그리기

```python
plt.figure(figsize=(12, 7))
plt.hist(value, label = ['1st', '2nd'], bins = 30) # bins : 연속적인 변수를 히스토그램으로 나타내기 위해 데이터를 나누는 개수
plt.grid(True)
plt.legend(bbox_to_anchor = (1.2, 1), loc='upper right')
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram 1')
```

```python
# 데이터 셋 별로 다른 색 입히기
plt.figure(figsize=(12, 6))
plt.hist(value, label = ['1st', '2nd'], bins = 30, color = ('yellow', 'green'))
plt.grid(True)
plt.legend(bbox_to_anchor = (1.2, 1), loc='upper right')
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram 2')
```

```python
# pyplot.hist() 함수를 이용한 히스토그램 그리기
value = np.random.standard_normal(500)
cm = plt.cm.get_cmap('Spectral')
plt.figure(figsize=(12, 6))

# Plot histogram
n, bins, patches = plt.hist(value, bins=30, color='green')
print("1. bins : " + str(bins))
print("2. the length of bins : " + str(len(bins)))
bin_centers = 0.5 * (bins[:-1] + bins[1:])
print("3. bin_centers : " + str(bin_centers))

# scale values to interval [0, 1] : 정규화 (normalization)
col = (bin_centers - min(bin_centers)) / (max(bin_centers) - min(bin_centers))
print("4. col : " + str(col))

for c, p in zip(col, patches):
    plt.setp(p, 'facecolor', cm(c))
    
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram 3')
plt.show()
```

- 정규화(Normalization)
    - 데이터의 전체 구간을 정해진 범위 안으로 변환하여 데이터를 관찰하는 방법
- 데이터를 정규화하는 방법
    - 최소-최대 정규화 : 0(최소값)과 1(최대값) 사이로 데이터의 범위를 변환
    - Z 점수 정규화 : 평균 및 표준 편차를 기반으로 데이터 조정.  
    데이터와 평균의 차이를 표준 편차로 나눈다.
    - 소수점 배열 : 특성 값의 소수점을 이동하여 데이터 크기 조정
- 최소-최대 정규화 수식
    - 새로운 값 = (현재 값 - 최소 값) / (최대 값 - 최소 값)

### Seaborn(Statistical Data Visualization)으로 시각화하기

#### Seaborn이란?
- Matplotlib를 기반으로 하는 Python 시각화 패키지
- 통계적 그래픽을 제공하는 고차원의 인터페이스 제공

#### "tips" 데이터로 boxplot 그리기
- seaborn 패키지 가져오기
- seaborn 스타일 옵션 설정
- seaborn 내장 데이터셋 "tips" 로드

```python
import seaborn as sns

# set aesthetic parameters in one step
sns.set(style="whitegrid", palette="pastel")

# Load the example tips dataset
data = sns.load_dataset("tips")
data
```

#### 요일(day)별 사례금(tips) 박스플롯 그리기

```python
sns.boxplot(x="day", y="tip", hue="sex", data=data, palette="muted")
```

#### "flights" 데이터로 heatmap 그리기

- seaborn 내장 데이터셋 "flights" 로드

```python
data2 = sns.load_dataset("flights")
data2
```

#### 데이터 pivot 시키기

`- index : "year"`  
`- column : "month"`  
`- value : "passengers"`

```python
pivoted_data = data2.pivot("year", "month", "passengers")
pivoted_data
```

- 스타일 옵션 변경하기
- Heatmap 으로 나타내기

```python
sns.set(context="poster", font="monospace")
sns.heatmap(pivoted_data, annot=True, fmt="d", linewidths=3)
```

- 옵션 초기화시키기

```python
sns.set()
sns.reset_orig() # reset params to defaults
sns.heatmap(pivoted_data, annot=True, fmt="d", linewidths=3)
```
